```python
prompt_user_edited = {
    1 : 'user edited post',
    0 : 'user did not edit post'
}
def prompt_caption(text):
    if text=='':
        return ''
    return 'caption is ' + text

def prompt_parameters_tone(text):
    if text=='':
        return ''
    return 'tone is ' + text

def prompt_parameters_chapter_title(text):
    if text=='':
        return ''
    return 'chapter title is ' + text
def promt_parameters_chapter_summary(text):
    if text=='':
        return ''
    return 'summary is ' + text

def promt_parameters_theme(text):
    if text=='':
        return ''
    return 'theme is ' + text

def prompt_parameters_photo_search_term(text):
    if text=='':
        return ''
    else:
        return 'search term used is ' + text

def prompt_has_logo(text):
    if text:
        return 'it has a logo'
    else:
        return 'it does not have a logo'
```

```python
import string


def remove_punctuation(text):
    return ''.join(char for char in text if char not in string.punctuation)

import re

def remove_special_characters_and_numbers(text):
    cleaned_text = re.sub(r'[^a-zA-Z\s]', '', text)
    return cleaned_text

import nltk
from nltk.corpus import stopwords

nltk.download('stopwords')

def remove_stopwords(text):
    words = text.split()
    stop_words = set(stopwords.words('english')) 
    filtered_words = [word for word in words if word.lower() not in stop_words]
    cleaned_text = ' '.join(filtered_words)
    return cleaned_text

contraction_mapping = {
    "ain't": "is not",
    "aren't": "are not",
    "can't": "cannot",
    "could've": "could have",
    "couldn't": "could not",
    "didn't": "did not",
    "doesn't": "does not",
    "don't": "do not",
    "hadn't": "had not",
    "hasn't": "has not",
    "haven't": "have not",
    "he'd": "he would",
    "he'll": "he will",
    "he's": "he is",
    "I'd": "I would",
    "I'll": "I will",
    "I'm": "I am",
    "I've": "I have",
    "isn't": "is not",
    "it'd": "it would",
    "it'll": "it will",
    "it's": "it is",
    "let's": "let us",
    "mightn't": "might not",
    "mustn't": "must not",
    "shan't": "shall not",
    "she'd": "she would",
    "she'll": "she will",
    "she's": "she is",
    "should've": "should have",
    "shouldn't": "should not",
    "that's": "that is",
    "there's": "there is",
    "they'd": "they would",
    "they'll": "they will",
    "they're": "they are",
    "they've": "they have",
    "wasn't": "was not",
    "we'd": "we would",
    "we'll": "we will",
    "we're": "we are",
    "we've": "we have",
    "weren't": "were not",
    "what'll": "what will",
    "what're": "what are",
    "what's": "what is",
    "what've": "what have",
    "where's": "where is",
    "who'd": "who would",
    "who'll": "who will",
    "who're": "who are",
    "who's": "who is",
    "who've": "who have",
    "won't": "will not",
    "wouldn't": "would not",
    "you'd": "you would",
    "you'll": "you will",
    "you're": "you are",
    "you've": "you have"
    # Add more contractions and their expansions as needed
}

def expand_contractions(text):
    # Split the text into words
    words = text.split()
    
    # Iterate through the words and replace contractions with their expansions
    expanded_words = [contraction_mapping.get(word, word) for word in words]
    
    # Reconstruct the text with expanded contractions
    expanded_text = ' '.join(expanded_words)
    
    return expanded_text

import re

def remove_urls(text):
    # Define a regular expression pattern to match URLs
    url_pattern = re.compile(r'https?://\S+|www\.\S+')
    
    # Use the pattern to replace URLs with an empty string
    cleaned_text = url_pattern.sub('', text)
    
    return cleaned_text
```

```python
def clean_text(text):
    text = text.lower()
    text = expand_contractions(text)
    text = remove_urls(text)
    text = remove_punctuation(text)
    text = remove_special_characters_and_numbers(text)
    text = remove_stopwords(text)
    return text
```

```python
clean_columns = ['caption','parameters_chapter_title','parameters_chapter_summary','parameters_tone', 'parameters_theme', 'parameters_photo_search_term']
```

```python
for col in clean_columns:
    training_data[col] = training_data[col].apply(lambda text : clean_text(str(text)))
```

```python
prompt_text = []
for i in range(len(training_data)):
    text = ''
    sample = training_data.iloc[i,:]
    text += prompt_user_edited[sample['user_edited']]
    text += ' '
    text += prompt_caption(sample['caption'])
    text += ' '
    text += prompt_parameters_chapter_title(sample['parameters_chapter_title'])
    text += ' '
    text += promt_parameters_chapter_summary(sample['parameters_chapter_summary'])
    text += ' '
    text += prompt_parameters_photo_search_term(sample['parameters_photo_search_term'])
    text += ' '
    text += prompt_parameters_tone(sample['parameters_tone'])
    text += ' '
    text += promt_parameters_theme(sample['parameters_theme'])
    text += ' '
    text += prompt_has_logo(sample['has_logo'])
    prompt_text.append(text)
```
